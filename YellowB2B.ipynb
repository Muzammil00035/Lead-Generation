```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
req = requests.get('https://www.yellowpages.net/listing/places/?bbox=&d=20&l=Karachi&lat=&lon=&q=cosmetics')
```

```python
req
```

```python
html = BeautifulSoup(req.text, 'html.parser')
```

```python
a = html.find_all('h2' , {'class': 'card__title'})
data = []
for i in a:
    rowdata = []
    new = i.find('a')
    text = new.getText()
    rowdata.append(text)
    link = new.get('href')
    
    
    another = requests.get(link)
       
    htmll = BeautifulSoup(another.text, 'html.parser')
    
    add = htmll.find_all('div' , {'class' : 'card__subtitle'})
    for k in add:
        
        adddiv = k.find('div')
        s= adddiv.find('span').text
        
        rowdata.append(s)
        
    web = htmll.find_all('div' , {'class' : 'company-header-www d-flex'})
    for k in web:
        
        webdiv = k.find('div')
        w= webdiv.find('a').get('href')
        
        rowdata.append(w)
        
    num = htmll.find_all('div' , {'class' : 'mdc-card__action-buttons'})
    for l in num:
        
        n= l.find('a').get('title')
        
        rowdata.append(n)
    
    
        
    
    
   
            
        
    
    
    
    
        
    data.append(rowdata)
    
    
df = pd.DataFrame(data)
df.columns = ['Company Name' , 'Address' , 'Website' , 'Phone']
df
```

```python
cs = df.to_csv('B2B.csv')
cs
```
